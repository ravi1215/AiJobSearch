```python
import gradio as gr
from langchain_community.document_loaders import WebBaseLoader
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import PromptTemplate
import pandas as pd
import uuid
import chromadb
from langchain_groq import ChatGroq

llm = ChatGroq(
    temperature=0,
    groq_api_key="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",
    model_name="llama-3.3-70b-versatile"
)

def preproces_job_posting(url, portfolio_csv):
    loader = WebBaseLoader(url)
    page_data = loader.load().pop().page_content
    prompt_extract = PromptTemplate.from_template("""
        ### SCRAPED TEXT FROM WEBSITE:
        {page_data}
        ### INSTRUCTION:
        The scraped text is from the career's page of a website.
        Your job is to extract the job postings and return them in JSON format containing the
        following keys: `role`, `experience`, `skills` and `description`.
        Only return the valid JSON.
        ### VALID JSON (NO PREAMBLE):
        """)
    chain_extract = prompt_extract | llm
    res_1 = chain_extract.invoke(input={'page_data': page_data})
    json_parser = JsonOutputParser()
    json_res = json_parser.parse(res_1.content)

    df = pd.read_csv(portfolio_csv)

    client = chromadb.PersistentClient('vectorstore')
    collections = client.get_or_create_collection(name="portfolio_app")

    if not collections.count():
        for i, row in df.iterrows():
            collections.add(documents=row['Resume'], ids=[str(uuid.uuid4())])

    job = json_res.get('skills', []) if type(json_res) == dict else json_res[0].get('skills', [])

    prompt_skills_and_question = PromptTemplate.from_template("""
        ### JOB DESCRIPTION:
        {job_description}

        ### INSTRUCTION:
        You are a 3rd-year Computer Science Engineering student at NSUT, passionate about exploring cutting-edge technologies in machine learning, deep learning, and natural language processing.
        Your expertise lies in understanding theoretical concepts and applying them to solve real-world problems effectively.
        Your job is to:
        1. Analyze the given job description to identify the required technical skills and match them with the provided skill set to calculate a percentage match.
        2. Generate a list of relevant interview questions based on the job description that align with industry expectations.
        3. Return the information in JSON format with the following keys:
            - `skills_match`: A dictionary where each key is a skill, and the value is the matching percentage.
            - `interview_questions`: A list of targeted questions derived from the job description to test practical understanding.

        Only return the valid JSON.
        ### VALID JSON (NO PREAMBLE):
        """)
    chain_skills_and_question = prompt_skills_and_question | llm
    res1 = chain_skills_and_question.invoke({"job_description": str(job)})
    final_result = json_parser.parse(res1.content)
    return final_result

def gradio_interface(url, portfolio_csv):
    try:
        result = preproces_job_posting(url, portfolio_csv)
        return result
    except Exception as e:
        return {"error": str(e)}

with gr.Blocks(
    theme=gr.themes.Monochrome(
        primary_hue="blue",
        secondary_hue="green",
        font="Roboto",
    )
) as app:
    gr.Markdown(
        """
        <div style="text-align:center; background-color: #1f2937; padding: 20px; border-radius: 10px;">
          <h1 style="color: #ffffff;">💼 Job Scraping & Interview Prep AI</h1>
          <p style="color: #d1d5db; font-size: 18px;">
          Revolutionize your job search with AI-powered job posting analysis and interview preparation! 🚀
          </p>
        </div>
        """,
        elem_id="header",
    )

    with gr.Row():
        with gr.Column(scale=3):
            gr.Image(value="/content/ian-schneider-TamMbr4okv4-unsplash.jpg", elem_id="job-image")
        with gr.Column(scale=2):
            url_input = gr.Textbox(
                label="🔗 Website URL",
                placeholder="Paste the job posting URL here",
                lines=1,
                elem_id="url-input"
            )
            portfolio_input = gr.File(label="📁 Upload Your Portfolio (CSV)")

    analyze_button = gr.Button(
        value="✨ Analyze Job Posting",
        elem_id="analyze-btn",
        variant="primary",
    )
    output_box = gr.JSON(label="🎯 Results")

    analyze_button.click(gradio_interface, inputs=[url_input, portfolio_input], outputs=output_box)

app.launch()
```
